```python
import numpy as np
import pandas as pd 
import os
import matplotlib.pyplot as plt
from tools import smape
from sklearn.decomposition import PCA
import urllib.request

import read_data
import json
```

```python
data=read_data.data()
```

```python
proteins = {}
for prot in data.protein_names:
    uf = urllib.request.urlopen('https://rest.uniprot.org/uniprotkb/'+prot+'.fasta')
    html = uf.read().decode("utf-8") 
    info = html.splitlines()[0]
    sequence = ''.join(html.splitlines()[1:])
    proteins[prot] = {'info': info, 'sequence': sequence}
```

```python
proteins
```

```python
with open('prot_seq.json', 'w') as fp:
    json.dump(proteins, fp)
```
